```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import roc_auc_score, roc_curve

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings("ignore")
```

```python
#Data Gathering
df = pd.read_csv("Fish.csv")
df
```

```python
# EDA
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
x = df.drop("Species",axis = 1)
y = df["Species"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = 18, stratify = y)
x_train
```

```python
y_test.info()
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
# Model Training
model = LogisticRegression(multi_class = "ovr")
model.fit(x_train,y_train)
```

```python
# Evaluation 
def metrics(a,b):
    cnf_matrix = confusion_matrix(a,b)
    print("The Confusion Matrix is :\n", cnf_matrix)
    print("*"*40)
    accuracy = accuracy_score(a,b)
    print("Accuracy is :",accuracy)
    print("*"*40)

    clf_report = classification_report(a,b)
    print("Classification Report :\n",clf_report)
    print("*"*40)


    
    
```

```python
# Evaluation for Testing Data 
y_pred = model.predict(x_test)
metrics(y_test,y_pred)
```

```python
# Evaluation on training data
y_pred_train = model.predict(x_train)
metrics(y_train,y_pred_train)
```
